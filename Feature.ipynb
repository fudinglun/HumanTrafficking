```python
import numpy as np
import pandas as pd
import pdb
import nltk
import string
from nltk.corpus import stopwords
import csv
from random import shuffle
```

```python
def clean_list_of_word(words, remove_stopping_word=True):
    """remove punctuations"""
    translator = str.maketrans('', '', string.punctuation)
    result = list(map(lambda x: x.translate(translator) , words))
    
    """remove stopping word"""
    if remove_stopping_word:
        result = [word for word in result if word not in stopwords.words('english')]
    return result
```

```python
df = pd.read_csv('../all_ter_data_dropna.csv')
one_list = list(df.loc[df['ht'] == 1.0]['Unnamed: 0'])
zero_list = list(df.loc[df['ht'] == 0.0]['Unnamed: 0'])
one_train = one_list[2000:16000]
zero_train = zero_list[2000:16000]
validation_list = one_list[:2000] + zero_list[:2000]
```

```python
shuffle(validation_list)
```

```python
def build_dictionary(ids):
    dic = {}
    for i in ids:
        row = df.loc[df['Unnamed: 0'] == i]
        details = clean_list_of_word(str(row['juicy_details'].item()).split(' '))
        for word in details:
            if word in dic:
                dic[word] += 1
            else:
                dic[word] = 1
    return dic
```

```python
one_dic = build_dictionary(one_train)
```

```python
zero_dic = build_dictionary(zero_train)
```

```python
diff = set(one_dic) - set(zero_dic)
```

```python
d = {}
for k in diff:
    d[k] = one_dic[k]
```

```python
import operator
sorted_x = sorted(d.items(), key=operator.itemgetter(1), reverse=True)
```
